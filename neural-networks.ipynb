##  **Building a Simple logistic Regressor**

```python
import numpy as np
np.random.seed(1)

# Now We will take the given values to start our logistic Regressor


X = np.array([[0,1,0],
              [1,0,0],
              [1,1,1],
              [0,1,1]])
y = np.array([[0,1,1,0]]).T
```

```python
#Defining Sigmoid and Loss Function

def sigmoid(x):
    return 1/(1+np.exp(-x))

def bce_loss(y,y_hat):
    N = y.shape[0]
    loss = -1/N * (y*np.log(y_hat) + (1 - y)*np.log(1-y_hat))
    return loss
```

```python
#Initialising Required Variables 
N = y.shape[0]
losses = []
alpha = 1
epochs = 20

#Initialising Parameters such that mean = 0 and SD =1
W = 2*np.random.random((3,1)) - 1
b = 0
```

```python
#Training Step

for i in range(epochs):
    #Forward pass
    z = X.dot(W) + b
    A = sigmoid(z)
    
    #Calculating Loss
    loss = bce_loss(y,A)
    losssum = np.sum(loss)
    print("Epoch :", i,"loss ->",losssum)
    losses.append(losssum)
    
    #Gradient Descent
    dz = (A - y)
    dW = 1/N * np.dot(X.T,dz)
    db = 1/N * np.sum(dz,axis=0,keepdims=True)
    
    
    #Changing Parameters
    W -= alpha * dW
    b -= alpha * db
    
```

```python
import matplotlib.pyplot as plt
plt.plot(losses)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

## Using Keras to build Neural networks

```python
from keras.layers import Dense, Activation
from keras.models import Sequential
```

```python
# First We Create an Empty Model with No layers
model = Sequential()
model.add(Dense(3,input_dim = 2))
model.add(Activation('tanh'))
model.add(Dense(1))
model.add(Activation('sigmoid'))
model.summary()
```

```python
model.compile(optimizer='sgd',loss='binary_crossentropy',metrics=['acc'])
```

```python
history = model.fit(X,y,epochs=900)
```
